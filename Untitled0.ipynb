```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Gerekli programları kullanmak için import girildi.

```python
# 'veri.csv' dosyasının okunması
df = pd.read_csv('/Student Performance new.csv')

# İlk birkaç satırın görüntülenmesi
print(df.head())
print(df.tail)
print(df.head)
```

Veri seti kod ile analiz etmeye hazırlandı.

```python
# Veri setinin genel bilgilerinin alınması
print(df.info())
print(df.columns)
print(df.shape)
print(df.dtypes)
# Veri setinin özet istatistikleri
print(df.describe())
```

Veri hakkında genel bilgiler kontrol edildi.

```python
# Eksik değerlerin sayısının kontrol edilmesi
print(df.isnull().sum())

# Eksik değerlerin doldurulması (örnek: ortalama ile doldurma)

```

Eksik değerler kontroll edildi ve doğrulandı

```python
# Örnek olarak bir sütun dağılım grafiği (histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['math percentage'], bins=30)
plt.title('Not Dağılımları')
plt.xlabel('Değerler')
plt.ylabel('Not')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['test preparation course'], bins=30)
plt.title('Hazırlık kursu testleri')
plt.xlabel('Değerler')
plt.ylabel('Test sayısı')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['reading score percentage'], bins=30)
plt.title('Okuma tablosu')
plt.xlabel('Değerler')
plt.ylabel('Not')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['writing score percentage'], bins=30)
plt.title('Yazma tablosu')
plt.xlabel('Değerler')
plt.ylabel('Not')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['parental level of education'], bins=30)
plt.title('Ebeveyn Eğitim Düzeyi tablosu')
plt.xlabel('Eğitim Dereceleri')
plt.ylabel('Sayı')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['race/ethnicity'], bins=30)
plt.title('Irk ve Kökenler')
plt.xlabel('Kökenler')
plt.ylabel('Sayı')
plt.show()
```

Bu aşamada artık veri görselleştirilmeye başlanıyor.

```python
# Korelasyon matrisinin hesaplanması
corr_matrix = df.corr(numeric_only=True)

# Korelasyon matrisinin görselleştirilmesi
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasyon Matrisi')
plt.show()
```

Korelasyon analizi yapılıyor. Ve görüldüğü üzere matematikte başarılı olma yüzdesinin çeşitli etkenlerle korelasyon içinde olduğu sonucuna varılabiliyor.
